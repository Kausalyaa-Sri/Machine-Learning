```python
import matplotlib.pyplot as plt
import math
```

```python
x=range(1,100)
y=[math.sin(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 8 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)#creating 2 height and 1 width and activating 1st graph
plt.plot(x,y_sin)
plt.title("sineeeeeeee")
plt.subplot(2,1,2)#creating 2 height and 1 width and activating 2nd graph
plt.plot(x,y_cos)
plt.show()
```
